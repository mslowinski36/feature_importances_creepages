```python
import pandas as pd 
import os 
import numpy as np
import matplotlib.pyplot as plt
```

```python
names=['time', 'velocity', 'long_creepage', 'yaw_angle', 
           'wear_number', 'lateral_displacement', 'lateral_creepage',
 'spin_creepage',"yaw_angle_pos","contact_angle","long_creep_force","lateral_creep_force"]

variables=['velocity', 'long_creepage', 'yaw_angle', 'lateral_displacement', 'lateral_creepage',
 'spin_creepage',"yaw_angle_pos","contact_angle","long_creep_force","lateral_creep_force"]
```

```python
podgorna_data = pd.read_csv('podgorna.txt', names=names)
podgorna_data = podgorna_data[10:]

for var in variables:

    fig, ax1 = plt.subplots()

    ax1.plot(podgorna_data['wear_number'], color='tab:blue')
    ax1.set_ylabel('wear_number', color='tab:blue')

    ax2 = ax1.twinx()  
    ax2.plot(podgorna_data[var], color='tab:orange')
    ax2.set_ylabel(var, color='tab:orange')

    plt.title('Podgorna: wear_number vs. {}'.format(var))
    plt.show()
```
