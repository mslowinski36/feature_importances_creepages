```python
import pandas as pd 
import os 
import numpy as np
import matplotlib.pyplot as plt
```

```python
podgorna_data = pd.read_csv(r'C:\Users\Marcin\Desktop\INFRAIL\algorytmy\feature importance - creepages\raw_data\podgorna.txt', names=['time', 'velocity', 'long_creepage', 'yaw_angle', 'wear_number', 'lateral_displacement', 'lateral_creepage',
                                                                                                                                      'spin_creepage',"yaw_angle_pos","contact_angle","long_creep_force","lateral_creep_force"])
podgorna_data = podgorna_data[10:]

fig, ax1 = plt.subplots()

ax1.plot(podgorna_data['wear_number'], color='tab:blue')
ax1.set_ylabel('wear_number', color='tab:blue')

ax2 = ax1.twinx()  
ax2.plot(podgorna_data['contact_angle'], color='tab:orange')
ax2.set_ylabel('contact_angle', color='tab:orange')

plt.title('Podgorna')
plt.show()
```

```python
plt.plot(podgorna_data['long_creepage'])
plt.plot(podgorna_data['lateral_creepage'])
```

```python
plt.plot(podgorna_data['long_creep_force'])
plt.plot(podgorna_data['lateral_creep_force'])
```

```python
baraniaka_data=pd.read_csv(r'C:\Users\Marcin\Desktop\INFRAIL\algorytmy\feature importance - creepages\raw_data\baraniaka.txt',names=['time','velocity','long_creepage','yaw_angle',"wear_number",'lateral_displacement',"lateral_creepage","spin_creepage","yaw_angle_pos","contact_angle","long_creep_force","lateral_creep_force"])
baraniaka_data=baraniaka_data[10:4900]
fig, ax1 = plt.subplots()

ax1.plot(baraniaka_data['wear_number'], color='tab:blue')
ax1.set_ylabel('wear_number', color='tab:blue')

ax2 = ax1.twinx()  
ax2.plot(baraniaka_data['yaw_angle'], color='tab:orange')
ax2.set_ylabel('Yaw Angle', color='tab:orange')

plt.title('Baraniaka')
plt.show()
```

```python
plt.plot(baraniaka_data['long_creep_force'])
plt.plot(baraniaka_data['lateral_creep_force'])
#plt.plot(baraniaka_data['long_creepage'])
#plt.plot(baraniaka_data['lateral_creepage'])
```

```python
awf_data=pd.read_csv(r'C:\Users\Marcin\Desktop\INFRAIL\algorytmy\feature importance - creepages\raw_data\awf.txt',names=['time','velocity','long_creepage','yaw_angle',"wear_number",'lateral_displacement',"lateral_creepage","spin_creepage","yaw_angle_pos","contact_angle","long_creep_force","lateral_creep_force"])
awf_data=awf_data[10:4900]

fig, ax1 = plt.subplots()

ax1.plot(awf_data['wear_number'], color='tab:blue')
ax1.set_ylabel('wear_number', color='tab:blue')

ax2 = ax1.twinx()  
ax2.plot(awf_data['yaw_angle'], color='tab:orange')
ax2.set_ylabel('Yaw Angle', color='tab:orange')

plt.title('AWF')
plt.show()
```

```python
#plt.plot(baraniaka_data['long_creep_force'])
#plt.plot(baraniaka_data['lateral_creep_force'])
plt.plot(awf_data['long_creepage'])
plt.plot(awf_data['lateral_creepage'])
```

```python
# Pocz wszystkie ramki danych
merged_data = pd.concat([podgorna_data, baraniaka_data, awf_data])

# Wymieszaj wiersze
merged_data = merged_data.sample(frac=1).reset_index(drop=True)
```

```python
merged_data.to_csv('dataset.csv',index=None)
```
